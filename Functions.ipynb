  For Example:

6. Explain the concept of generators in Python and how they are defined.
- In Python, a generator is a special type of function that is used to create an iterator in a simple and memory-efficient way. Unlike normal functions that use the return statement to send back a single value, generators use the yield keyword to produce a sequence of values one at a time. Each time the generator is called, it remembers its last execution state and resumes from there, instead of starting over.

  Generators are defined like normal functions, but instead of return, they contain one or more yield statements. When the function is called, it does not execute immediately; instead, it returns a generator object, which can be iterated over using a for loop or the next() function.

  Example:
        
      def count_up_to(n):
        i = 1
      while i <= n:
        yield i
        i += 1

      # Using the generator
      for num in count_up_to(5):
        print(num)

    # Output:
      1
      2
      3
      4
      5

7. What are the advantages of using generators over regular functions?
- In Python, generators have several advantages over regular functions when it comes to handling sequences of data.

  **Advantages of Generators over Regular Functions:**

  - **Memory Efficiency** – Generators produce one value at a time instead of storing the entire sequence in memory.

  - **Lazy Evaluation**– Values are generated only when needed, reducing unnecessary computation.

  - **Infinite Sequences** – Generators can handle unbounded data streams (e.g., sensor data, live feeds).

  - **Simpler Code** – Using yield, they automatically implement the iterator protocol, avoiding extra code.

  - **Better Performance** – Faster execution for large datasets since they do not load everything into memory.

  - **Improved Readability** – Generator functions make iterative logic easier to write and understand.

  - **Resource Management** – Generators stop execution after yielding values, reducing CPU and memory usage.

8. What is a lambda function in Python and when is it typically used?
- A lambda function in Python is a small, anonymous function that is defined using the keyword lambda instead of the usual def. It can take any number of arguments but can have only a single expression. The expression is automatically returned, so no return statement is needed.

  It's syntax is:

      lambda arguments: expression

  For example:

      square = lambda x: x * x
      print(square(5))   
      
      # Output: 25


  Here, lambda x: x * x creates an anonymous function that returns the square of a number.

  **When is it typically used?**

  **Short, simple functions** – When a function is only needed once or is very small.

  **Inside higher-order functions** – Often used with functions like map(), filter(), and sorted().

      nums = [1, 2, 3, 4, 5]
      squares = list(map(lambda x: x*x, nums))
      print(squares)   
      
      # Output:  [1, 4, 9, 16, 25]


  **As arguments** – Passed as inline functions without defining a full function using def.

  **Cleaner code** – Helps avoid unnecessary function definitions for simple operations.

9. Explain the purpose and usage of the `map()` function in Python.
- In Python, the map() function is used to apply a function to each item of an iterable (like a list, tuple, or set) and return a new map object (which is an iterator). It allows us to perform transformations on data without using explicit loops, making the code shorter and cleaner.

  It's syntax is:

      map(function, iterable)


  **function** → A function that defines the operation to perform.

  **iterable** → A sequence (list, tuple, etc.) whose elements are processed.

  Example: Using a normal function

      def square(x):
        return x * x

      numbers = [1, 2, 3, 4, 5]
      result = map(square, numbers)
      print(list(result))  
      
       # Output: [1, 4, 9, 16, 25]

  Example: Using a lambda function

      numbers = [1, 2, 3, 4, 5]
      result = map(lambda x: x + 10, numbers)
      print(list(result))   
      
      # Output: [11, 12, 13, 14, 15]

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- Difference between `map()`, `reduce()`, and `filter()` functions are:
    - **map()**

  **Purpose:** Applies a function to each element of an iterable and returns a new iterable (map object).

  **Usage:** Transformation of data.

  Example:

      numbers = [1, 2, 3, 4]
      result = map(lambda x: x*x, numbers)
      print(list(result))   
      
      # Output: [1, 4, 9, 16]

  - **filter()**
   
   **Purpose:** Filters elements from an iterable based on a condition (True/False) and returns only the elements that satisfy it.

  **Usage:** Selection of data.

  Example:

      numbers = [1, 2, 3, 4, 5]
      result = filter(lambda x: x % 2 == 0, numbers)
      print(list(result))   
      
      #Output: [2, 4]

    - **reduce()**

  **Purpose:** Applies a function cumulatively to the elements of an iterable, reducing it to a single value.

  **Usage:** Aggregation (sum, product, max, etc.).

  **Note:** reduce() is in the functools module.

  Example:

      from functools import reduce
      numbers = [1, 2, 3, 4]
      result = reduce(lambda x, y: x + y, numbers)
      print(result)   
      
      # Output: 10

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
# the list.

def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

nums = [10, 20, 3, 40, 5, 6, 7, 8]
print("Sum of even numbers:", sum_even_numbers(nums))
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
# terms.
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

fib = Fibonacci(10)
for num in fib:
    print(num, end=" ")
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for value in powers_of_two(5):
    print(value, end=" ")
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(CDS_Syllabus):
    """Generator function to read a file line by line with encoding fix"""
    try:
        with open(CDS_Syllabus, 'r', encoding='utf-8', errors='ignore') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error: File '{CDS_Syllabus}' not found.")

# Example usage
CDS_Syllabus = input("Enter the file name (with extension): ")

for line in read_file_lines(CDS_Syllabus):
    print(line)
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
tuples_list = [(1, 5), (3, 2), (4, 7), (2, 1)]

sorted_list = sorted(tuples_list, key=lambda x: x[1])

print("Original List:", tuples_list)
print("Sorted List (by 2nd element):", sorted_list)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius_temps = [0, 10, 20, 30, 40, 100]

def c_to_f(c):
    return (c * 9/5) + 32

fahrenheit_temps = list(map(c_to_f, celsius_temps))

print("Celsius:", celsius_temps)
print("Fahrenheit:", fahrenheit_temps)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
input_str = "Hello, Yogesh! Welcome to Python."

def not_vowel(char):
    return char.lower() not in 'aeiou'

result = ''.join(filter(not_vowel, input_str))

print("Original String:", input_str)
print("String without vowels:", result)
```
